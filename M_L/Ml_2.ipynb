```python
import numpy as np
import matplotlib.pyplot as plt

t= np.arange(0.0,2.0,0.01)

s= 1+ np.cos(2*np.pi*t)

plt.plot(t,s,'b')
plt.grid()
plt.xlabel('Time(t)')
plt.ylabel('Voltage(ev)')
plt.legend(['s'])
plt.title(" Cosin wave plot(Cos(x))")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(2,1,1)  # Subplot(row, col, postion)
plt.plot(x1,y1,'r')
plt.xlabel("X1")
plt.ylabel('Y1')
plt.legend(['cos(y1)'])
plt.title('Amp(y1)')
plt.show()

plt.subplot(2,1,1)
plt.plot(x2,y2,'g')
plt.xlabel("X2")
plt.ylabel('Y2')
plt.legend(['cos(y2)'])
plt.title('Amp(y2)')
plt.show()
```
